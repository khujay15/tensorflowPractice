```python
import pandas as pd
data = pd.read_csv('./datasetF.csv')
data.head(20)

import tensorflow as tf
X1 = tf.placeholder(tf.float32, [20,])
X2 = tf.placeholder(tf.float32, [20,])
X3 = tf.placeholder(tf.float32, [20,])
X4 = tf.placeholder(tf.float32, [20,])
X5 = tf.placeholder(tf.float32, [20,])
X6 = tf.placeholder(tf.float32, [20,])
X7 = tf.placeholder(tf.float32, [20,])
Y = tf.placeholder(tf.float32, [20,])

W1 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W2 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W3 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W4 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W5 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W6 = tf.Variable(tf.random_normal([20,], stddev=0.35))
W7 = tf.Variable(tf.random_normal([20,], stddev=0.35))


#W3 = tf.Variable(tf.random_normal([20,], stddev=0.35))

b1 = tf.Variable(tf.zeros([20,]))
name= data[data.columns[0]]
def current_rank(ranking):
    lis = []
    num=0
    for i in ranking:
        lis.append((i,name[num]))
        num+=1

    lis.sort()
    
    for i in lis:
        print (i)

data_1= data['조회수']
data_2= data['좋아요']
data_3= data['1차']
data_4= data['조회수2']
data_5= data['좋아요2']
data_6= data['위스플']
data_7= data['국적']
data_y= data['2차']
#Second = tf.matmul(W1, X1) + tf.matmul(W2,X2) + tf.matmul(W3,X3) + b1
Second =  W1*tf.log(X1) + W2*tf.log(X2) + W3*(X3) + b1+ W4*tf.log(X4) + W5*tf.log(X5)+ W6*(X6)  + W7*(X7) 
#cost = tf.reduce_mean((Second - Y))
cost = tf.reduce_mean(tf.sqrt(tf.pow((Second - Y),2.0)))
train =  tf.train.GradientDescentOptimizer(0.1).minimize(cost)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

#summary_writter = tf.summary.FileWriter('./tensorboard',sess.graph)

for i in range(4000):
    _, rr= sess.run([train,Second], feed_dict= { X1: data_1, X2: data_2, X3: data_3, X4: data_4,X5: data_5, X6: data_6,X7: data_7,Y: data_y})
    if i ==1 :
        current_rank(rr)
        print()
    if i == 100 or i==1000 :
        current_rank(rr)
        _, cc= sess.run([train,Second], feed_dict= { X1: data_1, X2: data_2, X3: data_3, X4: data_4,X5: data_5, X6: data_6,X7: data_7,Y: data_y})
        print ("cc: %s"%cc)
        print()
        
ranking = sess.run([Second], feed_dict= { X1: data_1, X2: data_2, X3: data_3, X4: data_4,X5: data_5, X6: data_6,X7: data_7,Y: data_y})
```
